## 1- Veri Setinin Hazırlanması: NLTK'den movie_reviews veri setini kullanacağız.
## 2- Özellik Çıkarma: TF-IDF kullanarak metinleri sayısal özelliklere dönüştüreceğiz.
## 3- Optimum Küme Sayısının Belirlenmesi: Elbow yöntemi ile küme sayısını belirleyeceğiz.
## 4- Kümeleme Performansının Değerlendirilmesi: Silhouette Score ile kümelerin kalitesini ölçeceğiz.
## 5- Kümeleme: KMeans algoritması ile kümeleme yapacağız.

```python
import nltk
from nltk.corpus import movie_reviews
import random
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
# 1. Veri Setini Hazırlama
nltk.download('movie_reviews')
documents = [(" ".join(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]
random.shuffle(documents)
```

```python
texts = [doc for doc, _ in documents]  # Yorum metinleri
labels = [label for _, label in documents]  # Etiketler (kullanılmayacak)
```

```python
# 2. Özellik Çıkarma (TF-IDF Vektörleştirme)
vectorizer = TfidfVectorizer(max_features=1000)  # TF-IDF ile vektörleştir
X = vectorizer.fit_transform(texts)
```

---------
## Elbow Grafiği ile optimum küme sayısının tespit edilmesi

```python
# 3. Optimum Küme Sayısını Belirleme (Elbow Yöntemi)
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, n_init = 10, random_state=42)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)
```

```python
# Elbow Grafiği
plt.figure(figsize=(8, 5))
plt.plot(k_range, sse, marker='o')
plt.xlabel("Küme Sayısı (k)")
plt.ylabel("SSE (Sum of Squared Errors)")
plt.title("Elbow Yöntemi ile Optimum Küme Sayısının Belirlenmesi")
plt.show()
```

```python
# 4. Kümeleme Performansının Değerlendirilmesi (Silhouette Score)
silhouette_scores = []
for k in range(2, 11):  # Silhouette score, 2 veya daha fazla kümelerde anlamlıdır
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)  # n_init parametresini manuel olarak ayarladık
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(score)
```

```python
# Silhouette Score Grafiği
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel("Küme Sayısı (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score ile Kümeleme Performansının Değerlendirilmesi")
plt.show()
```

```python
# 5. KMeans ile Kümeleme
optimal_k = 2  # Elbow veya Silhouette Score grafiğinden belirlediğiniz değeri kullanabilirsiniz
kmeans = KMeans(n_clusters=optimal_k, n_init = 10, random_state=42)
clusters = kmeans.fit_predict(X)
```

```python
# Kümelerden bazı örnekleri görüntüleyelim
for i in range(optimal_k):
    print(f"\nKüme {i + 1}:")
    cluster_texts = [texts[j] for j in range(len(clusters)) if clusters[j] == i]
    print("Örnek Yorumlar:", cluster_texts[:3])  # Her kümeden ilk 3 örnek
```

## Görselleştirmeler

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def plot_word_clouds(texts, clusters, num_clusters):
    plt.figure(figsize=(15, 10))
    for i in range(num_clusters):
        cluster_texts = " ".join([texts[j] for j in range(len(clusters)) if clusters[j] == i])
        wordcloud = WordCloud(width=800, height=400, background_color="white", max_words=50).generate(cluster_texts)
        
        plt.subplot(1, num_clusters, i + 1)
        plt.imshow(wordcloud, interpolation="bilinear")
        plt.axis("off")
        plt.title(f"Küme {i + 1}")
    plt.show()

# Küme sayısını belirtin (örneğin, 2 küme)
plot_word_clouds(texts, clusters, num_clusters=2)
```

```python
import pandas as pd
import seaborn as sns

# Küme dağılımı
cluster_counts = pd.Series(clusters).value_counts().sort_index()
plt.figure(figsize=(8, 6))
sns.barplot(x=cluster_counts.index, y=cluster_counts.values, palette="viridis")
plt.xlabel("Küme")
plt.ylabel("Yorum Sayısı")
plt.title("Her Kümede Bulunan Yorum Sayısı")
plt.show()
```

```python
from sklearn.decomposition import PCA

# PCA ile 2 boyuta indirgeme
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X.toarray())

# Görselleştirme
plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette="viridis", s=50)
plt.xlabel("PCA Bileşeni 1")
plt.ylabel("PCA Bileşeni 2")
plt.title("Kümeleme Sonuçlarının 2 Boyutta Görselleştirilmesi")
plt.legend(title="Kümeler")
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Her kümede en sık geçen kelimeleri bulmak için CountVectorizer kullanıyoruz
def get_top_words_per_cluster(texts, clusters, num_words=10):
    cluster_data = pd.DataFrame({'text': texts, 'cluster': clusters})
    top_words = {}

    for cluster in set(clusters):
        cluster_texts = cluster_data[cluster_data['cluster'] == cluster]['text']
        vectorizer = CountVectorizer(max_features=100, stop_words='english')
        X_cluster = vectorizer.fit_transform(cluster_texts)
        word_counts = X_cluster.toarray().sum(axis=0)
        top_words[cluster] = [word for word, count in sorted(zip(vectorizer.get_feature_names_out(), word_counts), key=lambda x: x[1], reverse=True)[:num_words]]

    return top_words

# KMeans ile kümeleme sonucunda elde ettiğimiz kümeleri kullanarak en sık kelimeleri çıkarıyoruz
top_words = get_top_words_per_cluster(texts, clusters)

# Her kümeden birkaç örnek yorum gösterecek görselleştirme
num_examples_per_cluster = 3
for cluster in set(clusters):
    print(f"\nKüme {cluster + 1}:")
    print("En Sık Geçen Kelimeler:", ", ".join(top_words[cluster]))
    
    # Her kümeden num_examples_per_cluster kadar örnek gösterelim
    cluster_texts = [texts[i] for i in range(len(clusters)) if clusters[i] == cluster]
    for example in cluster_texts[:num_examples_per_cluster]:
        print("\nÖrnek Yorum:\n", example[:200], "...")  # Yorumların sadece ilk 200 karakterini gösteriyoruz
    print("-" * 50)
```
