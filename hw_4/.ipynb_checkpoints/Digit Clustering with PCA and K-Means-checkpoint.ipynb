# 1. MNIST Veri Setinin ve Kütüphanelerin yüklenmesi
## MNIST veri setini yükleyip, görüntüler üzerinde normalizasyon işlemi uyguluyoruz.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from tensorflow.keras.datasets import mnist

# MNIST veri setini yükleme
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Eğitim verisini kullanacağız, görüntüleri 2D'den 1D'ye dönüştürelim
x_train_flat = x_train.reshape(x_train.shape[0], -1)  # 28x28'i 784 boyutlu bir vektöre dönüştür
x_train_flat = x_train_flat / 255.0  # Normalizasyon

# Eğitim veri setini yeniden bölme (örneğin %80 eğitim, %20 test)
from sklearn.model_selection import train_test_split

x_train_new, x_valid, y_train_new, y_valid = train_test_split(
    x_train_flat, y_train, test_size=0.2, random_state=42
)

# Boyut bilgisi
print(f"Yeni Eğitim veri seti boyutu: {x_train_new.shape}")
print(f"Yeni Doğrulama veri seti boyutu: {x_valid.shape}")
```

------
## Görselleştirmeler

```python
# Rastgele 16 örnek görüntüyü çizdirme
plt.figure(figsize=(10, 10))
for i in range(16):
    plt.subplot(4, 4, i + 1)
    plt.imshow(x_train[i], cmap="gray")
    plt.axis("off")
    plt.title(f"Label: {y_train[i]}")
plt.suptitle("Orijinal MNIST Görselleri", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Piksel yoğunluğu histogramı
plt.figure(figsize=(8, 6))
plt.hist(x_train_flat.ravel(), bins=50, color="blue", alpha=0.7, edgecolor="black")
plt.title("Piksel Yoğunluğu Histogramı")
plt.xlabel("Piksel Değeri")
plt.ylabel("Frekans")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Eğitim ve doğrulama setinden örnek görseller
plt.figure(figsize=(12, 6))

# Eğitim verisinden 10 örnek
for i in range(5):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train_new[i].reshape(28, 28), cmap="gray")
    plt.axis("off")
    plt.title("Eğitim")

# Doğrulama verisinden 10 örnek
for i in range(5):
    plt.subplot(2, 5, i + 6)
    plt.imshow(x_valid[i].reshape(28, 28), cmap="gray")
    plt.axis("off")
    plt.title("Doğrulama")

plt.suptitle("Eğitim ve Doğrulama Verilerinden Örnekler", fontsize=16)
plt.tight_layout()
plt.show()
```

## Optimum Küme Sayısını Belirleme (Elbow Yöntemi)

```python
# Elbow yöntemi ile optimum küme sayısını belirleme
inertia = []
k_values = range(2, 11)  # Küme sayısı için 2'den 10'a kadar deneme yapılır

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x_train_new)
    inertia.append(kmeans.inertia_)

# Elbow yöntemi grafiği
plt.figure(figsize=(8, 6))
plt.plot(k_values, inertia, marker='o', linestyle='-', color='b')
plt.title("Elbow Yöntemi ile Optimum Küme Sayısı")
plt.xlabel("Küme Sayısı (k)")
plt.ylabel("Inertia")
plt.grid()
plt.show()
```

## Silhouette Skoru ile Kümeleme Performansını Değerlendirme
### Silhouette skoru, kümeleme performansını değerlendirir. En yüksek silhouette skoruna sahip küme sayısı optimum olabilir.

```python
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x_train_new)
    score = silhouette_score(x_train_new, kmeans.labels_)
    silhouette_scores.append(score)

# Silhouette skoru grafiği
plt.figure(figsize=(8, 6))
plt.plot(k_values, silhouette_scores, marker='o', linestyle='-', color='g')
plt.title("Silhouette Skorları")
plt.xlabel("Küme Sayısı (k)")
plt.ylabel("Silhouette Skoru")
plt.grid()
plt.show()
```

## PCA ile 2D Kümeleme Görselleştirme
### Veriyi 2D'ye indirgemek için PCA kullanarak kümeleme sonuçlarını görselleştiriyoruz.

```python
# Veriyi PCA ile 2D'ye indirgeme
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train_new)

# K-Means ile kümeleme (en iyi k değerini seçin, örneğin k = 10)
optimal_k = 10  # Elbow veya silhouette yöntemi ile belirlenen küme sayısı
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(x_train_pca)

# Kümeleme sonuçlarını görselleştirme
plt.figure(figsize=(10, 8))
for cluster in range(optimal_k):
    plt.scatter(
        x_train_pca[clusters == cluster, 0],
        x_train_pca[clusters == cluster, 1],
        label=f"Küme {cluster}"
    )
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c="red", label="Merkezler")
plt.title("Kümeleme Sonuçları (PCA ile 2D Görselleştirme)")
plt.xlabel("PCA Bileşen 1")
plt.ylabel("PCA Bileşen 2")
plt.legend()
plt.show()
```

```python
# Her kümede kaç örnek var?
unique, counts = np.unique(clusters, return_counts=True)

plt.figure(figsize=(10, 6))
plt.bar(unique, counts, color="skyblue")
plt.title("Kümelerdeki Örnek Sayıları")
plt.xlabel("Küme")
plt.ylabel("Örnek Sayısı")
plt.xticks(unique)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Veriyi PCA ile 3D'ye indirgeme
pca_3d = PCA(n_components=3)
x_train_pca_3d = pca_3d.fit_transform(x_train_new)

# 3D Kümeleme grafiği
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(
    x_train_pca_3d[:, 0], x_train_pca_3d[:, 1], x_train_pca_3d[:, 2],
    c=clusters, cmap="viridis", s=5, alpha=0.6
)
ax.set_title("Kümeleme Sonuçları (3D PCA)")
ax.set_xlabel("PCA 1")
ax.set_ylabel("PCA 2")
ax.set_zlabel("PCA 3")
plt.colorbar(scatter, ax=ax, label="Küme")
plt.show()
```

```python
# Her kümeden rastgele 10 örneği görselleştirme
num_samples_per_cluster = 10  # Her kümeden görselleştirilecek örnek sayısı
clusters_unique = np.unique(clusters)  # Tüm kümeleri al
plt.figure(figsize=(15, len(clusters_unique) * 2))

for cluster_id in clusters_unique:
    # Belirli küme için indeksleri bul
    cluster_indices = np.where(clusters == cluster_id)[0]
    # Rastgele örnekler seç
    random_indices = np.random.choice(cluster_indices, num_samples_per_cluster, replace=False)
    
    for i, idx in enumerate(random_indices):
        plt.subplot(len(clusters_unique), num_samples_per_cluster, cluster_id * num_samples_per_cluster + i + 1)
        plt.imshow(x_train_new[idx].reshape(28, 28), cmap="gray")
        plt.axis("off")
        if i == 0:
            plt.ylabel(f"Küme {cluster_id}", fontsize=12)

plt.suptitle("Kümeleme Sonuçları: Her Kümeden Rastgele Örnekler", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Her kümeden rastgele 10 örneği görselleştirme
num_samples_per_cluster = 10  # Her kümeden görselleştirilecek örnek sayısı
clusters_unique = np.unique(clusters)  # Tüm kümeleri al
plt.figure(figsize=(15, len(clusters_unique) * 2))

for cluster_id in clusters_unique:
    # Belirli küme için indeksleri bul
    cluster_indices = np.where(clusters == cluster_id)[0]
    # Rastgele örnekler seç
    random_indices = np.random.choice(cluster_indices, num_samples_per_cluster, replace=False)
    
    for i, idx in enumerate(random_indices):
        plt.subplot(len(clusters_unique), num_samples_per_cluster, cluster_id * num_samples_per_cluster + i + 1)
        plt.imshow(x_train_new[idx].reshape(28, 28), cmap="gray")
        plt.axis("off")
        plt.title(f"Label: {y_train_new[idx]}")  # Gerçek etiket gösterimi
        if i == 0:
            plt.ylabel(f"Küme {cluster_id}", fontsize=12)

plt.suptitle("Kümeleme Sonuçları: Her Kümeden Rastgele Örnekler ve Etiketler", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Validasyon verisini başlangıçta kullanılan PCA modeline göre indirgeme
x_valid_pca = PCA(n_components=2).fit_transform(x_valid)  # PCA, 2 bileşen

# Validasyon setinde kümeleri tahmin et
valid_clusters = kmeans.predict(x_valid_pca)

# Kümeler ile gerçek etiketleri eşleştirme
from scipy.stats import mode

cluster_to_label = {}  # Küme -> Gerçek Etiket eşleştirmesi

for cluster_id in np.unique(valid_clusters):
    # Kümenin indekslerini al
    cluster_indices = np.where(valid_clusters == cluster_id)[0]
    # Bu kümeye ait gerçek etiketleri bul
    cluster_labels = y_valid[cluster_indices]
    # En sık görülen gerçek etiketi belirle (mod)
    most_common_label = mode(cluster_labels).mode[0]
    cluster_to_label[cluster_id] = most_common_label

# Küme ve etiket eşleşmesini yazdırma
print("Küme -> Gerçek Etiket Eşleşmeleri:")
for cluster, label in cluster_to_label.items():
    print(f"Küme {cluster} -> Gerçek Etiket {label}")

# Validasyon setinde doğruluk hesaplama
correct_predictions = 0

for idx, cluster in enumerate(valid_clusters):
    predicted_label = cluster_to_label[cluster]
    true_label = y_valid[idx]
    if predicted_label == true_label:
        correct_predictions += 1

accuracy = correct_predictions / len(y_valid)
print(f"Validasyon Setinde Doğruluk: {accuracy * 100:.2f}%")
```
